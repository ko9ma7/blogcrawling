```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm_notebook
```

```python
path = './output/크롤링/통합/'
file_list = os.listdir(path)
```

```python
filter_word = pd.read_excel('filter_word.xlsx')
filter_word = filter_word['filter_word'].values
```

```python
#filter로 인해 사라지는 개수확인
befor_ls = []
after_ls = []
for name in tqdm_notebook(file_list):
    file = pd.read_csv(path + name)
    file = file.fillna('')
    
    filtered = file.full_text.apply(lambda x : any(ele in x for ele in filter_word))
    clean_file = file[~filtered]
    filtered = clean_file.title.apply(lambda x : any(ele in x for ele in filter_word))
    clean_file = clean_file[~filtered]
    
    after_ls.append(clean_file.shape[0])
    befor_ls.append(file.shape[0])
result = pd.DataFrame({'befor_ls':befor_ls,'after_ls':after_ls})
```
